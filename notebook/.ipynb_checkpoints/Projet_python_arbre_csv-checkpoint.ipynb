### ● Quel Maire a planté le plus d’arbres à Grenoble ?

```python
def arbres_plantes(annees):
    """
    Renvoie le nombre d'arbres plantés sur une période donnée
    annees : liste d'annees sur lesquelles compter
    """
    compte = 0
    for an in annees:
        if str(an) in arbres_par_an:
            compte += arbres_par_an[str(an)]
    return compte
```

```python
Maires = {'Mistral' : [i for i in range(1919,1932)],
          'Martin' : [i for i in range(1932,1935)]+[i for i in range(1945,1947)]+[i for i in range(1949,1959)],
          'Cocat' : [i for i in range(1935,1944)],
          'Lafleur' : [i for i in range(1944,1945)],
          'Bally' : [i for i in range(1947,1948)],
          'Perinetti' : [i for i in range(1948,1949)],
          'Michallon' : [i for i in range(1959,1965)],
          'Dubedout' : [i for i in range(1965,1983)],
          'Carignon' : [i for i in range(1983,1995)],
          'Destot' : [i for i in range(1995,2014)],
          'Piolle' : [i for i in range(2014,2020)]
         }
```

```python
max_compte = 0
Maires_updated = dict(Maires)
for Maire in Maires:
    compte = arbres_plantes(Maires[Maire])
    print("%s a planté %d arbres entre %d et %d"\
          %(str(Maire), compte, Maires[Maire][0], Maires[Maire][-1]))
    Maires_updated[Maire] = {'période':Maires[Maire],'nbr_arbres':compte}
    if compte > max_compte:
        max_compte = compte
        Maire_max_Arbres = str(Maire)
print("\nCONCLUSION :\nLe maire ayant planté le plus d'arbres (%d) est Mr. %s (%d - %d)"\
      %(max_compte, Maire_max_Arbres, Maires[Maire_max_Arbres][0], Maires[Maire_max_Arbres][-1]))
# using the newly generated dictionnary that contains names, years and trees of each mayor:
Maire_vert = max(Maires_updated, key=lambda item: item[1])
print(f"Checking that {Maire_vert} is indeed the one with most planted tree :\
{Maires_updated[Maire_vert]['nbr_arbres']}")
```

### ●	Comment la pratique de plantation a-t-elle évolué au cours du temps ? Diversifie-t-on plus maintenant qu’avant ? Vos analyses sont-elles en accord avec le communiqué de presse ? 

```python
# On va chercher à représenter l'évolution du nombre de genre, du nombre d'espèces,
# de la population maximale par genre, 
# de la population maximale par espèce, au cours du temps.

# Pour commencer, on crée une liste "clean" de dictionnaires, chaque dictionnaire représentant un arbre
# dont on connait l'année de plantation :
csv_list_clean = [row for row in csv_list if row["ANNEEDEPLANTATION"] != '']
csv_list_clean = sorted(csv_list_clean, key=lambda row:row["ANNEEDEPLANTATION"]) # cette ligne ordonne la liste selon l'année de plantation
```

```python
# Ensuite, on crée un dictionnaire plus simple : les clés sont les années de plantation
# les valeurs sont des dictionaires : le genre et l'espece de chaque arbre planté à cette année
Dict_evolution = dict()
for row in csv_list_clean:
    an = row["ANNEEDEPLANTATION"]
    if an in Dict_evolution:
        Dict_evolution[an]['genre'] += [row["GENRE_BOTA"]]
        Dict_evolution[an]['espece'] += [row["GENRE_BOTA"] + ' ' + row["ESPECE"]]
    else:
        Dict_evolution[an] = {'genre': [row["GENRE_BOTA"]],
                              'espece' : [row["GENRE_BOTA"] + ' ' + row["ESPECE"]]
                             };
```

```python
Dict_evolution['1924']
```

```python
# En utilisant la méthode Counter, on accède au nombre d'arbres par genre/espèce pour une année donnée:
Counter(Dict_evolution['2006']['genre']) #Counter(Dict_evolution['2006']['espece'])
```

```python
# On peut finalement mettre à jour le dictionnaire avec n'importe quelle information statistique par année
Dict_evolution_stat = dict(Dict_evolution)
for an in Dict_evolution_stat.keys():
    # Ici, on s'intéresse au genre botanique le plus planté chaque année:
    max_genre = max(Counter(Dict_evolution[an]['genre']), key=Counter(Dict_evolution[an]['genre']).get)
    # à hauteur de combien d'arbres:
    nbr_repr_max_genre_an = Counter(Dict_evolution[an]['genre'])[max_genre]
    # Puis quelle proportion des arbres plantés cette année ce genre représente :
    proportion_max_genre = nbr_repr_max_genre_an / sum(Counter(Dict_evolution[an]['genre']).values()) * 100
    # Et enfin combien de genres botaniques différents sont plantés annuellement
    genres = len(set(Counter(Dict_evolution[an]['genre'])))
    
    # De même sur les espèces
    max_espece = max(Counter(Dict_evolution[an]['espece']), key=Counter(Dict_evolution[an]['espece']).get)
    nbr_repr_max_espece_an = Counter(Dict_evolution[an]['espece'])[max_espece]
    proportion_max_espece = nbr_repr_max_espece_an / sum(Counter(Dict_evolution[an]['espece']).values()) * 100 
    especes = len(set(Counter(Dict_evolution[an]['espece'])))
    
    # Tous les résultats sont rajoutés au dictionnaire, associés à la clé = an:
    Dict_evolution_stat[an].update({'genre_max': max_genre,
                               'arbres_genre_max': nbr_repr_max_genre_an,
                               'genres_differents': genres,
                               'proportion_arbres_genre_max': proportion_max_genre,
                               'espece_max': max_espece,
                               'arbres_espece_max': nbr_repr_max_espece_an,
                               'especes_differentes': especes,
                               'proportion_arbres_espece_max': proportion_max_espece
                              })
    Dict_evolution_stat[an]['genre'] = Counter(Dict_evolution[an]['genre'])
    Dict_evolution_stat[an]['espece'] = Counter(Dict_evolution[an]['espece'])
                                          
```

```python
# On visualise l'evolution des parts de l'espèce/du genre la/le plus planté/e chaque année
fig, axes = plt.subplots(1,2, figsize=(10,4))
to_plot = {'dict_key':['proportion_arbres_genre_max', 'proportion_arbres_espece_max'],
           'title':['Part du genre botanique le + planté', """Part de l'espèce la + plantée"""]
          }
for i,ax in enumerate(axes):
    ax.bar([int(an) for an in Dict_evolution_stat.keys()], 
            [Dict_evolution_stat[an][to_plot['dict_key'][i]] for an in list_of_years],
            width=0.8,
            edgecolor='k',
            facecolor='b'
           )
    if i==0:
        ax.set_ylabel("% du total d'arbre\nplantés par an")
        ax.hlines(20,
                  xmin=int(min(Dict_evolution_stat.keys()))-1,
                  xmax=int(max(Dict_evolution_stat.keys()))+1,
                  colors='r'
                 )
    if i==1:
        ax.hlines(10,
                  xmin=int(min(Dict_evolution_stat.keys()))-1,
                  xmax=int(max(Dict_evolution_stat.keys()))+1,
                  colors='r',
                  linewidth= 1.5
                 )
    ax.set_xlabel("Année")
    ax.set_title(to_plot['title'][i])
    #ax.set_xlim(1950)
plt.tight_layout()
plt.show()
```

```python
def diversite_parc(an):
    """
    Renvoie la proportion du genre le plus présent et de l'espèce la plus présente dans le parc,
    ainsi que le nom du genre et le nom de l'espèce
    an : int, année à laquelle évaluer la population du parc 
    """
    counter_espece = Counter()
    counter_genre = Counter()
    for annee in Dict_evolution_stat.keys():
        if int(annee) <= an:
            counter_espece = counter_espece + Counter(Dict_evolution_stat[annee]['espece'])
            counter_genre = counter_genre + Counter(Dict_evolution_stat[annee]['genre'])
    max_espece = max(counter_espece,
                     key=counter_espece.get)
    nbr_repr_max_espece_an = Counter(counter_espece)[max_espece]
    prop_esp = nbr_repr_max_espece_an / sum(counter_espece.values()) * 100
    
    max_genre = max(counter_genre,
                     key=counter_genre.get)
    nbr_repr_max_genre_an = Counter(counter_genre)[max_genre]
    prop_genre = nbr_repr_max_genre_an / sum(counter_genre.values()) * 100
    
    return prop_genre, prop_esp, max_genre, max_espece
```
